```python
#Bibliotecas Utilizadas

import matplotlib.pyplot as plt
import math

#CIRCUITO RC PARALELO COM R em SÉRIE
#Valor de R e C da associação em paralelo:
Rp = input ('Digite R assoc. paralela [Ohm] = ')
Rp = float (Rp)
Cp = input ('Digite C assoc. paralela [micro Farad] = ')
Cp = float (Cp)
#Valor do R série:
Rs = input ('Digite R assoc. série [Ohm] = ')
Rs = float (Rs)

#atribuindo valor para a frequencia inicial:
expoente_inic = input ('Expoente inicial para a frequência de base 10: ')
print('')
valor_exp = float (expoente_inic)
freq_in = 10**valor_exp

print (f'Resistor Paralelo: {Rp} Ohm')
print (f'Capacitor Paralelo: {Cp} micro Farad')
print (f'Resistor Série: {Rs} Ohm')
print (f'Frequencia inicial: {freq_in} Hz')
Cp = Cp*10**(-6)

print('-------------------------------------------------------------------------------------------------------------')
#iniciando a varredura 

list_real = []
list_imag = []
list_freq = []

while valor_exp <= 5:
    valor_exp = valor_exp + 0.1
    freq = 10**(valor_exp - 0.1)
    #Calculando a magnitude da Impedância complexa
    G = 1/Rp
    B = 2 * math.pi * freq * Cp
    Mod_Y = ((G)**2 + (B)**2)**(1/2)
    Mod_Z = 1 / Mod_Y
    #Calculando o ângulo da Impedância complexa 
    radteta = math.atan(B/G)
    
    #Parte real e parte imaginária da impedância
    zreal = (Mod_Z * math.cos(radteta)) + Rs
    zimag = Mod_Z * math.sin(radteta)
    
    print ('Re[Z*]= {:.3f} Ohm, Imag[Z*]= {:.3f} Ohm,para a frequencia nominal:{:.3f}Hz'.format(zreal, zimag, freq))
    list_real.append(zreal)
    list_imag.append(zimag)
    list_freq.append(freq)
    print ('')
    print ('---------------------------------------------------------------------------------------------------------')


plt.plot(list_freq, list_real,'b')
plt.plot(list_freq, list_imag, 'y')
plt.xscale("log")
plt.ylabel('Re[Z*] e Imag[Z*] ohm')
plt.xlabel('Frequencia Hz')
plt.axis ([min(list_freq), max(list_freq), -200, max(list_real) +100])
plt.show()
print ('Azul: Re[Z*]')
print ('Amarelo: Imag[Z*]')
```
